```python
import torch
import matplotlib.pyplot as plt
```

```python
#单个分类器，四分类100epoch
train_acuracys_single= torch.load('train_acuracys_single.pt')
test_acuracys_single= torch.load('test_acuracys_single.pt')
print(train_acuracys_single)
print(test_acuracys_single)
print(max(train_acuracys_single))
print(max(test_acuracys_single))
```

```python
plt.figure()
plt.plot(range(len(train_acuracys_single)), train_acuracys_single, marker='o', color='black', linewidth=1.5, label='Train Acc')
plt.plot(range(len(test_acuracys_single)), test_acuracys_single, marker='o', color='purple', linewidth=1.5, label='Test Acc')

y_value = 95.57142857142857
plt.axhline(y=y_value, color='black', linestyle='--')
plt.text(len(train_acuracys_single)-1, y_value+2, f'TrainAcc = {y_value:.2f}', color='black', va='bottom')

y_value2 = 94.58333333333333
plt.axhline(y=y_value2, color='purple', linestyle='--')
plt.text(len(train_acuracys_single)-1, y_value2-4, f'TestAcc = {y_value2:.2f}', color='purple', va='bottom')

plt.legend()
plt.xlabel('Num_Epoch')
plt.ylabel('Accuracy(%)')
plt.grid(False)
plt.show()
```

```python
#3个弱分类器集成，每个5个epoch，四分类问题
#train_acuracys= torch.load('train_acuracys.pt')
#test_acuracys= torch.load('test_acuracys.pt')
train_final_acuracys= torch.load('train_final_acuracys.pt')
test_final_acuracys= torch.load('test_final_acuracys.pt')
train_ensemble_accuracys = torch.load('train_ensemble_accuracys.pt')
test_ensemble_accuracys = torch.load('test_ensemble_accuracys.pt')
print(train_final_acuracys)
print(test_final_acuracys)
print(train_ensemble_accuracys)
print(test_ensemble_accuracys)
```

```python
train_final_acuracys4 = train_final_acuracys[:4]
test_final_acuracys4 = test_final_acuracys[:4]
train_ensemble_accuracys4 = train_ensemble_accuracys[:4]
test_ensemble_acuracys4 = test_ensemble_accuracys[:4]
```

```python
train_ensemble_accuracys5 = [x * 100 for x in train_ensemble_accuracys4]
test_ensemble_acuracys5 = [y * 100 for y in test_ensemble_acuracys4]

```

```python
print(train_final_acuracys4)
print(test_final_acuracys4)
print(train_ensemble_accuracys5)
print(test_ensemble_acuracys5)
```

```python
plt.figure()
plt.plot(range(len(train_final_acuracys4)), train_final_acuracys4, marker='x', color='black', linewidth=1.5, label='Train Acc')
plt.plot(range(len(test_final_acuracys4)), test_final_acuracys4, marker='x', color='purple', linewidth=1.5, label='Test Acc')
plt.plot(range(len(train_ensemble_accuracys5)),train_ensemble_accuracys5, marker='o', color='red', linewidth=1.5, label='Ensemble Train Acc')
plt.plot(range(len(test_ensemble_acuracys5)),test_ensemble_acuracys5, marker='o', color='blue', linewidth=1.5, label='Ensemble Test Acc')
# 添加每个数据点的y值标记
for i, y in enumerate(train_final_acuracys4):
    plt.text(i, y, f'{y:.2f}', color='black', ha='right', va='bottom')
for i, y in enumerate(test_final_acuracys4):
    plt.text(i, y, f'{y:.2f}', color='purple', ha='right', va='bottom')
for i, y in enumerate(train_ensemble_accuracys5):
    plt.text(i, y, f'{y:.2f}', color='red', ha='right', va='bottom')
for i, y in enumerate(test_ensemble_acuracys5):
    plt.text(i, y, f'{y:.2f}', color='blue', ha='right', va='bottom')
    
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=2)

plt.xlabel('Num_Classifiers')
plt.ylabel('Accuracy(%)')
plt.grid(False)
plt.show()
```
